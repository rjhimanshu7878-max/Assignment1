1. Write a Python script that accepts a number from the command line 
and checks whether it is prime. Explain how command-line arguments 
are accessed.


```python
import sys

def prime(n):
    """Check if a number is prime."""
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

if __name__ == "__main__":
    if len(sys.argv) != 2:
        print("Usage: python script.py <number>")
    else:
        try:
            num = int(sys.argv[1])
            if prime(num):
                print(f"{num} is a prime number.")
            else:
                print(f"{num} is not a prime number.")
        except ValueError:
            print("Please enter a valid integer.")
```

2.Write a program using different data types to demonstrate the difference between ‘is’ and ‘==’. Explain the output.


```python
# Integer
a = 10
b = 10
print("Equal operator in Integer:-",a == b)   
print("is operator in Integer",a is b)   

# List
x = [1, 2, 3]
y = [1, 2, 3]
print("Equal operator in List:-",x == y)  
print("is operator in List",x is y)   

# String
s1 = "hello"
s2 = "hello"
print("Equal operator in String:-",s1 == s2)
print("is operator in String:-",s1 is s2)
```

3. Write a program that iterates through a list, skips negative numbers, 
stops execution at zero, and handles non-integer values using exception 
handling.

```python
data = [10, -5, 3, "abc", 7, 0, 4, -2]
for item in data:
    try:
        num = int(item)
        if num < 0:
            continue        
        if num == 0:
            print("Zero encountered. Stopping execution.")
            break          
        print("Valid number:", num)

    except ValueError:
        print("Non-integer value skipped:", item)
```

5. Write a user-defined function that returns the largest and smallest 
values from a list without using built-in functions.

```python
def largest_smallest(lst):
    for i in range(len(lst)):
        for j in range(len(lst)):
            if lst[i] < lst[j]:
                lst[i], lst[j] = lst[j], lst[i]

    smallest = lst[0]
    largest = lst[-1]

    return largest, smallest


# Example
nums = [34, 12, 56, 7, 89, 23]

largest, smallest = largest_smallest(nums)

print("Largest:", largest)
print("Smallest:", smallest)
```
